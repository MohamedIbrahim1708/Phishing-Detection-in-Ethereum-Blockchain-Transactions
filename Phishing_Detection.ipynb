```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Load the dataset
df = pd.read_csv('transaction_dataset.csv').drop(columns=['Unnamed: 0', 'Index'])
```

```python
# Define categorical column names with correct leading spaces
categorical_cols = [' ERC20 most sent token type', ' ERC20_most_rec_token_type']
```

```python
# Encode categorical variables
df_encoded = pd.get_dummies(df, columns=categorical_cols)
```

```python
# Ensure all columns are numeric after encoding
X = df_encoded.drop(columns=['FLAG', 'Address'])  # Exclude non-numeric and target columns
y = df['FLAG']
```

```python
# Check for NaN values
X.isna().sum()
```

```python
# Impute NaN values with the mean of the column
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
# Balance the dataset
oversampler = SMOTE(sampling_strategy='auto')
undersampler = RandomUnderSampler(sampling_strategy='auto')
```

```python
X_resampled, y_resampled = oversampler.fit_resample(X_imputed, y)
X_resampled, y_resampled = undersampler.fit_resample(X_resampled, y_resampled)
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
```

```python
# Splitting into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Print out the shapes of the resulting datasets to verify
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
import time
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Helper function to evaluate and display results
def evaluate_model(model, X_train, X_test, y_train, y_test):
    start_time = time.time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    execution_time = time.time() - start_time
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'execution_time': execution_time,
        'confusion_matrix': cm
    }

# Initialize models
models = {
    'KNN': KNeighborsClassifier(),
    'Decision Tree (Entropy)': DecisionTreeClassifier(criterion='entropy'),
    'Decision Tree (Gini)': DecisionTreeClassifier(criterion='gini'),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

```python
# Function to apply PCA
def apply_pca(X_train, X_test, n_components):
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    return X_train_pca, X_test_pca, pca
```

```python
# Function to apply Decision Tree for feature selection
def apply_decision_tree_fs(X_train, y_train, X_test, max_features):
    tree_clf = DecisionTreeClassifier()
    tree_clf.fit(X_train, y_train)
    feature_importances = tree_clf.feature_importances_
    indices = np.argsort(feature_importances)[::-1][:max_features]
    return X_train[:, indices], X_test[:, indices], indices
```

```python
# Define feature selection methods
feature_selection_methods = {
    'PCA': apply_pca,
    'Decision Tree': apply_decision_tree_fs
}
```

```python
# Number of components/features to select
n_components = 20
max_features = 20
```

```python
# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Evaluate models with feature selection
results = {}
selected_features = {}
```

```python
for fs_method_name, fs_method in feature_selection_methods.items():
    print(f"\nApplying feature selection using {fs_method_name}...\n")
    
    if fs_method_name == 'PCA':
        X_train_fs, X_test_fs, pca = fs_method(X_train_scaled, X_test_scaled, n_components=n_components)
        # Get feature names by examining the loadings of the components
        pca_components = pca.components_
        feature_contributions = np.abs(pca_components).sum(axis=0)
        feature_indices = np.argsort(feature_contributions)[::-1][:n_components]
        feature_names = X.columns[feature_indices]
    elif fs_method_name == 'Decision Tree':
        X_train_fs, X_test_fs, indices = fs_method(X_train_scaled, y_train, X_test_scaled, max_features=max_features)
        feature_names = X.columns[indices]
    
    selected_features[fs_method_name] = feature_names
    print(f"Selected features for {fs_method_name}: {feature_names}")
    
    for model_name, model in models.items():
        print(f"Evaluating {model_name} with {fs_method_name}...")
        results[f"{model_name} ({fs_method_name})"] = evaluate_model(model, X_train_fs, X_test_fs, y_train, y_test)
        print(f"Results for {model_name} with {fs_method_name}:")
        print(f"Accuracy: {results[f'{model_name} ({fs_method_name})']['accuracy']}")
        print(f"Precision: {results[f'{model_name} ({fs_method_name})']['precision']}")
        print(f"Recall: {results[f'{model_name} ({fs_method_name})']['recall']}")
        print(f"Execution Time: {results[f'{model_name} ({fs_method_name})']['execution_time']}")
        print("\n")
```

```python
# Print summary of results
print("Summary of Results:")
for model_fs_name, metrics in results.items():
    print(f"{model_fs_name}:")
    for metric_name, metric_value in metrics.items():
        if metric_name != 'confusion_matrix':
            print(f"  {metric_name}: {metric_value}")
    print("\n")

# Print selected features for each method
print("Selected Features:")
for fs_method_name, features in selected_features.items():
    print(f"{fs_method_name}: {features}")
```
